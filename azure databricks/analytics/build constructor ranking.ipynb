
## **Sources**
1. results_session_fact
2. constructor_dim

##Output Columns
1. season
2. onstructor_id
3. constructor_name
4. nationality
5. race starts
6. total_points
7. number_of_wins
8. number_of_podium
9. standing_position

```sql
create or replace view formular1.gold.constructor_standing_view as

with constructor_session as 
(
    select  r.season,
        r.constructor_id,
        c.constructor_name,
        c.nationality,
        count(*) as race_starts,
        sum(r.points) as total_points,
        count_if(r.is_win) as total_wins,
        count_if(r.has_podium) as number_of_podium

from 
formular1.gold.results_session_fact r
join 
formular1.gold.gold_constructors_dim c
on r.constructor_id = c.constructor_id

group by  r.season,
        r.constructor_id,
        c.constructor_name,
        c.nationality
)

select *,
        rank() over(partition by season order by total_points desc) as standings
from constructor_session
```
