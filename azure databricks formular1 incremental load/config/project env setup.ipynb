# project environment setup for formula1 project
- Create external 'location databricks-formula1-ext-loc-formular1-incr'
- create catalog formular1_incr

###Access Cloud Storage


```sql
%fs ls abfss://formular1-incr@databricksstorageext101.dfs.core.windows.net/
```

### Create External Location

```sql
%sql

CREATE EXTERNAL LOCATION IF NOT EXISTS `databricks_course_ext_101_formular1_incr` 
URL 'abfss://formular1-incr@databricksstorageext101.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `databricks-course-sc`)
COMMENT 'external location for formular1-incr container'
```


### create catalog formular1 incr


```sql
show catalogs
```

```sql
create catalog if not exists formular1_incr
managed location 'abfss://formular1-incr@databricksstorageext101.dfs.core.windows.net/'
comment 'create catalog for formular1-incr project'
```


### create schemas landing, bronze, silver, gold

```sql
create schema if not exists formular1_incr.landing;
create schema if not exists formular1_incr.bronze
managed location 'abfss://formular1-incr@databricksstorageext101.dfs.core.windows.net/bronze';
create schema if not exists formular1_incr.silver
managed location 'abfss://formular1-incr@databricksstorageext101.dfs.core.windows.net/silver';
create schema if not exists formular1_incr.gold
managed location 'abfss://formular1-incr@databricksstorageext101.dfs.core.windows.net/gold'
comment 'bronze schema created'
```
