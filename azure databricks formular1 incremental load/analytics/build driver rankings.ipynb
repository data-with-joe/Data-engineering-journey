
## **Sources**
1. results_session_fact
2. dim_drivers

##Output Columns
1. season
2. driver_id
3. driver_name
4. nationality
5. race starts
6. total_points
7. number_of_wins
8. number_of_podium
9. standing_positions

```sql
select *  from 
 formular1.gold.results_session_fact r
 join formular1.gold.gold_drivers_dim d
 on r.driver_id = d.driver_id
```

```sql
create or replace view formular1.gold.drivers_standing_view as
 with session_ranking as 
 (
    select r.season,
        r.driver_id,
        d.driver_name,
        d.nationality,
        count(*) as race_start,
        sum(r.points) as total_point,
        count_if(r.is_win) as total_wins,
        count_if(r.has_podium) as has_podium
 from 
 formular1.gold.results_session_fact r
 join formular1.gold.gold_drivers_dim d
 on r.driver_id = d.driver_id
 group by r.season,
        r.driver_id,
        d.driver_name,
        d.nationality
)

select *,
        rank() over(partition by season order by total_point desc, total_wins desc ) as standings
       
from session_ranking
```

```sql
select * from drivers_standing_view
where season = '2022'
```
