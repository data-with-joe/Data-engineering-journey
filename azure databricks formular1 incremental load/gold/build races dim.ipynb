
1. read silver races table
2. read silver circuits table
3. join both tables using circuit_id
4. selected required collumn for BI
5. write data to gold dim races table


```python
%run ../config/config-env
```

```python
silver_race_table = f'{catalog_name}.{silver_schema}.races'
silver_circuits_table = f'{catalog_name}.{silver_schema}.circuits'
target_table = f'{catalog_name}.{gold_schema}.dim_races'
```

```python
races_df = spark.read.table(silver_race_table)
circuits_df = spark.read.table(silver_circuits_table)
```

```python

from pyspark.sql import functions as F
dim_races_df =( 
               races_df.join(circuits_df, races_df.circuit_id == circuits_df.circuit_id , "inner")
               
        ).select(
            races_df.season,
            races_df.round,
            races_df.race_name,
            races_df.race_date,
            circuits_df.circuit_name,
            circuits_df.locality,
            circuits_df.country
        )
```

```python
display(dim_races_df)
```

```python
(
    dim_races_df
        .write
        .format('delta')
        .mode('overwrite')
        .saveAsTable(target_table)
)
```
