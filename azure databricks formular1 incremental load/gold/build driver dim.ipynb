```python
driver_df = f'{catalog_name}.{silver_schema}.drivers'
ref_region_df = f'{catalog_name}.{gold_schema}.ref_nationality_region'
target_table = f'{catalog_name}.{gold_schema}.gold_drivers_dim'
```

```python
driver_read_df =spark.read.table(driver_df)
ref_region_read_df =spark.read.table(ref_region_df)
```

```python
joined_df = driver_read_df.join(
            ref_region_read_df, driver_read_df.nationality == 
                                ref_region_read_df.nationality,
                                'left'
).select(
    driver_read_df.driver_id,
    driver_read_df.driver_name,
    driver_read_df.date_of_birth,
    driver_read_df.nationality,
    ref_region_read_df.region
)
```

```python
(
    joined_df
        .write
        .format('delta')
        .mode('overwrite')
        .saveAsTable(target_table)
)
```
